```python
#import h5py
import pandas as pd
hdf = pd.HDFStore(r'C:\Users\PICHAU\Downloads\madrid.h5', 'r')

chaves=hdf.keys()
print(chaves)
    
```

```python
#visualizando o conteúdo dentro de cada key

for i in range(len(chaves)):
    print('------- Chave {} -------\n'.format(chaves[i]))
    print(hdf[chaves[i]].head())
    print('------------------------------------------------\n')
```

```python
#inserindo os dados da key master dentro do df

df = hdf['master']
df.head()
```

```python
import folium
from folium.plugins import HeatMap
#Criando o mapa (lat e long de madrid)
m = folium.Map(location=[df.lat[0], df.lon[0]], zoom_start=11, tiles='Stamen Terrain') #

x=[]
for i in range(len(df)):
    folium.Marker([df['lat'][i], df['lon'][i]], popup=df['address'][i], tooltip=df['name'][i]).add_to(m)
    folium.CircleMarker([df['lat'][i], df['lon'][i]], popup=df['name'][i], radius=2,color='blue').add_to(m)
    calor=df['lat'][i], df['lon'][i]
    x.append(calor)

#mapa de calor, é preciso fazer uma lista com as lats e longs conforme acima (calor)
HeatMap(x, radius=30).add_to(m)
```
